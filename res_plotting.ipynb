```python
import numpy as np
import pandas as pd
```

```python
total_loss = []
reconstruction_loss = []
kl_loss = []
with open("result/200epoch.txt") as mytxt:
    for line in mytxt:
        line = line.replace('\n','')
        splitted = line.split(",")
        if len(splitted)>1:
            ind_tot = splitted[0].find(":")
            total_loss.append(float(splitted[0][ind_tot+2:]))
            ind_rescons = splitted[1].find(":")
            reconstruction_loss.append(float(splitted[1][ind_rescons+2:]))
            ind_kl = splitted[2].find(":")
            kl_loss.append(float(splitted[2][ind_kl+2:]))
```

```python
import matplotlib.pyplot as plt
plt.plot(total_loss,alpha=0.6)

plt.plot(reconstruction_loss,alpha=0.6)

plt.plot(kl_loss,alpha=0.6)
plt.ylabel('Loss')
plt.xlabel('Iterations')
plt.title('Loss vs. Batch Iterations')
plt.legend(['Total loss','Reconstructed loss','KL loss'])
plt.show()
```
